```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def Get_Histgram(im):
    
    plt.subplot(2, 1, 1)
    a = np.array(im)
    plt.imshow(a)
    plt.gray()
    plt.axis('off')
    
    plt.subplot(2, 1, 2)
    plt.hist(a.flatten(), bins=np.arange(256 + 1))
    
    plt.show()
    
def GaussianNoise(src):
    
    row,col= src.shape
    mean = 0
    sigma = 15
    gauss = np.random.normal(mean,sigma,(row,col))
    gauss = gauss.reshape(row,col)
    noisy = src + gauss

    return noisy


def RayleighNoise(src):
    
    row,col= src.shape
    mean = 30
    rayleigh = np.random.rayleigh(mean,(row,col))
    rayleigh = rayleigh.reshape(row,col)
    noisy = src + rayleigh

    return noisy


def GammaNoise(src):
    
    row,col= src.shape
    mean = 2
    sigma = 20
    gamma = np.random.gamma(mean,sigma,(row,col))
    gamma = gamma.reshape(row,col)
    noisy = src + gamma

    return noisy

    
img = cv2.imread('test-pattern.tif', 0)
Get_Histgram(img)

Gnoise = GaussianNoise(img)
Get_Histgram(Gnoise)

Rnoise = RayleighNoise(img)
Get_Histgram(Rnoise)

Gammanoise = GammaNoise(img)
Get_Histgram(Gammanoise)
```
